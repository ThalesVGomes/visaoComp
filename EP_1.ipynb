```python
imgs = io.imread_collection("*.jpg")
```

```python
io.imshow(imgs[0])
```

```python
n_colunas, n_linhas = 5, 2

fig, axs = plt.subplots(n_linhas, n_colunas, figsize=(20, 20))
num_fotos = n_colunas * n_linhas
for pos, objeto in enumerate(imgs[:num_fotos]):
    l, c = (pos // n_colunas, pos % n_colunas)
    axs[l, c].imshow(objeto)
    axs[l, c].axis('off')
plt.tight_layout()
plt.show()
```
