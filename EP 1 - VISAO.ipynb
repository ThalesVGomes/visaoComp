```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from glob import glob
%matplotlib inline
```

```python
path = r'C:\Users\Thales\Desktop\Fotos Visao'
```

```python
files = glob(path + '\*.jpg') # Isso assume que não há subpastas, apenas imagens dentro de uma pasta única
```

```python
def get_images(files, rows, cols):
    fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(16,10))
    for imp, ax in zip(files, axes.ravel()):
        img = mpimg.imread(imp)
        ax.imshow(img)
        ax.axis('off')
```

```python
get_images(files, 5, 5)
```
